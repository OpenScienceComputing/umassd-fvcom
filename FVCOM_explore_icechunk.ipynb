# FVCOM Visualization (Modern Method)
* Open STAC item
* Get asset to Virtual Icechunk which references AWS Open Data NetCDF files
* Open icechunk asset in xarray using xpystac
* Wrap xarray as xugrid dataset (use UGRID conventions)
* Visualize using trimesh from hvplot.xugrid accessor

```python
import xarray as xr
import pystac
```

## Get virtual Icechunk asset from STAC Catalog

```python
catalog = pystac.Catalog.from_file("https://umassd-fvcom.s3.amazonaws.com/gom3/hindcast/stac/catalog.json")
item = catalog.get_item(id='umassd-fvcom-gom3-hindcast@J07NZVPPBK3FVTWWSN1G')
assets = item.get_assets(media_type="application/vnd.zarr+icechunk", role="latest-version")
```

```python
asset = list(assets.values())[0]
asset
```

## Open Icechunk assset in xarray

```python
import xpystac
ds_ugrid = xr.open_dataset(asset, chunks={}, engine="stac")
```

```python
ds_ugrid
```

## Create xugrid dataset
Xugrid is a package largely developed by Deltares that understands and makes use of the UGRID conventions for unstructured grid data. 

```python
import xugrid as xu
```

```python
uds = xu.UgridDataset(ds_ugrid)
```

#### Visualize using hvplot.xugrid accessor

```python
import hvplot.xugrid  # Requires https://github.com/holoviz/hvplot/pull/1724
```

#### node-based variable

```python
uds['temp'].attrs['location']
```

```python
da = uds['temp'].sel(time='2012-01-01 12:00', method='nearest').isel(siglay=0).load()
da.hvplot.trimesh(rasterize=True, geo=True, tiles='OSM', cmap='turbo',
              colorbar=True, frame_width=600)
```

#### face-based variable

```python
uds['u'].attrs['location']
```

```python
da = uds['u'].sel(time='2012-01-01 12:00', method='nearest').isel(siglay=0).load()
da.hvplot.trimesh(rasterize=True, geo=True, tiles='OSM', cmap='turbo',
              colorbar=True, frame_width=600)
```
